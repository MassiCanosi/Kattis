## GlitchBot

```python
X,Y = input().split(" ")
destination = []
destination.append(int(X))
destination.append(int(Y))

n = int(input())
instructions = []

for instruction in range(n):
    instructions.append(input())

#####################

def move(player_direction, position):
    match player_direction:
        case 'UP':
            position[1] += 1
        case 'DOWN':
            position[1] -= 1
        case 'RIGHT':
            position[0] += 1
        case 'LEFT':
            position[0] -=1
    
    return position


def turn(step, player_direction):

    if step == 'Right':

        match player_direction:
            case 'UP':
                player_direction = 'RIGHT'
            case 'DOWN':
                player_direction = 'LEFT'
            case 'RIGHT':
                player_direction = 'DOWN'
            case 'LEFT':
                player_direction = 'UP'

    elif step == 'Left':

        match player_direction:
            case 'UP':
                player_direction = 'LEFT'
            case 'DOWN':
                player_direction = 'RIGHT'
            case 'RIGHT':
                player_direction = 'UP'
            case 'LEFT':
                player_direction = 'DOWN'

    return player_direction


#####################

distinct_instructions = ['Forward', 'Right', 'Left']
loop_destination = [1000,1000]

different = True

#####################

while different:

    for i, step in enumerate(instructions):

        instruction_adj = instructions.copy()

        for di in distinct_instructions:

            if di != step:
                instruction_adj[i] = di 
            else:
                continue


            position = [0,0]
            player_direction = 'UP'

            for step in instruction_adj:

                if step == 'Forward':

                    position = move(player_direction, position)
                    

                elif step in ['Right', 'Left']:

                    player_direction = turn(step, player_direction)
            
            loop_destination = position

            if position == destination:
                different = False
                break
            
        if not different:
            break

#####################

for i,v in enumerate(instructions):
    if v != instruction_adj[i]:
        
        index = i+1
        value = instruction_adj[i]

print(index, value)
```
